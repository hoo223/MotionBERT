```python
labels = []
for i in range(len(data['annotations'])):
    labels.append(data['annotations'][i]['label'])
```

```python
total = 0
data_dict = {}
for i in tqdm(range(120)):
    if i not in data_dict.keys():
        data_dict[i] = {'train' : [], 'val' : []}
    action_idxs = np.where(np.array(labels)==i)[0]
    #print(i, len(action_idxs))
    total += len(action_idxs)
    sub_list = []
    for idx in action_idxs:
        frame_dir = data['annotations'][idx]['frame_dir']
        subject = frame_dir[:4]
        if frame_dir in data['split']['xsub_train']:
            data_dict[i]['train'].append(subject)
        else:
            data_dict[i]['val'].append(subject)
    data_dict[i]['train'] = list(set(data_dict[i]['train']))
    data_dict[i]['val'] = list(set(data_dict[i]['val']))
    #break
```
