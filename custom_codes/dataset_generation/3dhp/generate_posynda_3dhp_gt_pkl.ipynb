```python
root_3dhp = '/home/hrai/Datasets/HAAI/3DHP'
source = 'posynda' # 
test_data_path = os.path.join(root_3dhp, source, 'data_test_3dhp_ori.npz')
test_data = np.load(test_data_path, allow_pickle=True)['data'].item()
```

```python
save_folder = '/home/hrai/codes/MotionBERT/data/motion3d/3dhp_gt_test_pkl'
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

subject_list = test_data.keys()
for subject in subject_list:
    data = {}
    for key in ['joint_2d', 'confidence', 'camera_name', 'action', 'source', 'frame', 'cam_3d']:
        data[key] = []

    #print(test_data[subject].keys())
    joint_2d = mpi_inf_3dhp2h36m(test_data[subject]['data_2d'])
    cam_3d = mpi_inf_3dhp2h36m(test_data[subject]['data_3d'])
    print(joint_2d.shape, cam_3d.shape)
    print(cam_3d[0])
    frame_num = [i for i in range(joint_2d.shape[0])]
    source = [subject]*len(frame_num)
    action = [subject]*len(frame_num)
    camera_id = [subject]*len(frame_num)

    # store
    data['joint_2d'] = np.array(joint_2d).copy()
    data['confidence'] = np.ones([len(joint_2d), 17, 1])
    data['camera_name'] = np.array(camera_id).copy()
    data['action'] = np.array(action).copy()
    data['source'] = np.array(source).copy()
    data['frame'] = np.array(frame_num).copy()
    data['cam_3d'] = np.array(cam_3d).copy()

    file_name = subject + '.pkl'
    save_path = os.path.join(save_folder, file_name)
    savepkl(data, save_path)
```
